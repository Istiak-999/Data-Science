```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Gender = le.fit_transform(df.Gender)
df
```

```python
dummies = pd.get_dummies(df.Profession)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(["Profession","Doctor"], axis='columns')
final
```

```python
train=final.drop(["Annual Income ($)"],axis='columns')
test=final["Annual Income ($)"]
                 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train,test,test_size=0.3) 
```

```python
X_test
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

# decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
re=DecisionTreeRegressor()
```

```python
re.fit(X_train, y_train)
```

```python
predict=re.predict(X_test)
```

```python
re.score(X_test, y_test)
```

```python
plt.scatter(y_test, predict)
#plt.plot(y_test, predict,color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted Duration')
plt.show()
```

# random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
pr=rf.predict(X_test)
```

```python
rf.score(X_test, y_test)
```

```python
plt.scatter(y_test, pr)
#plt.plot(y_test, predict,color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted Duration')
plt.show()
```
